```python
import numpy as np
import matplotlib.pyplot as plt
from ga import ga
```

## Create Example: $4x^2$

```python
length = 11
x = np.zeros(length)
y = np.zeros_like(x)
for i in range(1, len(x)):
    x[i] = x[i-1] + 1/(len(x)-1)  
    y[i] = 4*x[i]**2

plt.figure(figsize=(15, 5))
plt.plot(x, y)
plt.title("Function to find best fit to: $4x^2$")
plt.show()
```

## Initialise Genetic Algorithm Fitter

```python
pop_size = 50000
max_iters = 10000
term_threshold = 1e-5
keep_next_iter = 0.2
remove_next_iter = 0.2
mutation_rate = 0.5
ga_fitter = ga(pop_size, max_iters, term_threshold, keep_next_iter, remove_next_iter, mutation_rate)  
```

## Obtain optimal parameter

```python
eqn = lambda a, z: a*z**2
constraints = np.zeros((1, 2))
constraints[0, 0] = 1e-6
constraints[0, 1] = 1e2
params = ga_fitter.fit(y, [x], eqn, constraints, plot=True, save_path="x_2")
print(params)
```
